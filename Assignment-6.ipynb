# Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)

```python
import numpy as np


```

```python
M = float(input("Enter a value for M: "))
x_values = np.linspace(-10, 10, num=100)  # Generates 100 points from -10 to 10

print("Generated x values:", x_values)
```

```python
y1 = M * x_values**2  # y = M * x^2
y2 = M * np.sin(x_values)  # y = M * sin(x)

print("Computed y1 values (M * x^2):", y1)
print("Computed y2 values (M * sin(x)):", y2)
```

#  Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x_values, y1, 'r--', label=r'$y = M \cdot x^2$')  # Red dashed line
plt.plot(x_values, y2, 'b-', label=r'$y = M \cdot \sin(x)$')  # Blue solid line

plt.legend()
plt.grid(True)
plt.title("Plot of $y = Mx^2$ and $y = Msin(x)$")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()
```

# Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
subjects = ["Math", "Science", "English", "History", "Computer"]
scores = [85, 92, 78, 88, 95]
```

```python
import pandas as pd

df = pd.DataFrame({"Subject": subjects, "Score": scores})
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
colors = sns.color_palette("husl", len(subjects))  # Generates different colors

ax = sns.barplot(x="Subject", y="Score", data=df, palette=colors)

# Annotate each bar with its respective score
for i, score in enumerate(scores):
    ax.text(i, score + 1, str(score), ha='center', fontsize=12, fontweight='bold')

plt.title("Scores of Subjects", fontsize=14)
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
import numpy as np

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)


data = np.random.randn(50)

print("Generated Data:", data)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(data), 'g-', linewidth=2)
plt.title("Cumulative Sum Plot")
plt.xlabel("Index")
plt.ylabel("Cumulative Sum")
plt.grid(True)

plt.show()
```

```python
x_values = np.arange(50)

plt.figure(figsize=(8, 5))
plt.scatter(x_values, data, c='b', alpha=0.6, edgecolors='black')
plt.title("Scatter Plot with Random Noise")
plt.xlabel("Index")
plt.ylabel("Random Values")
plt.grid(True)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))


axes[0, 0].plot(np.cumsum(data), 'g-', linewidth=2)
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

axes[0, 1].scatter(x_values, data, c='b', alpha=0.6, edgecolors='black')
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)

axes[1, 0].axis("off")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

# Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
url = "https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv"
data = pd.read_csv(url)

# Convert 'total_profit' to numerical value
data['total_profit'] = pd.to_numeric(data['total_profit'], errors='coerce')

# Line plot for total profit
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
# Multiline plot for product sales
plt.figure(figsize=(12, 8))

for column in data.columns[1:-1]:  # Exclude 'month_number' and 'total_profit'
    sns.lineplot(x='month_number', y=column, data=data, label=column, marker='o')

plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar plot for all attributes
data_bar = data.drop(columns='month_number')  # Exclude 'month_number' for clarity

data_bar.plot(kind='bar', figsize=(14, 8))
plt.title('Bar Chart for All Features/Attributes')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
